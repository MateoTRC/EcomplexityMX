```python
complexity_data_mex_pairs_socioecon=pd.read_csv('../output_data/complexity_data_mex_pairs_socioecon.csv')
complexity_data_mex_pairs_socioecon
```

```python

Estados=complexity_data_mex_pairs_socioecon['location_code'].unique()

years_pairs=complexity_data_mex_pairs_socioecon['year_pair'].unique()
```

```python
product=112511

year_pair='2004_2019'

complexity_data_year=complexity_data_mex_pairs_socioecon.loc[complexity_data_mex_pairs_socioecon['year_pair']==year_pair]

complexity_data_year.loc[complexity_data_year['Product']==prod]

complexity_data_mex_pairs_socioecon.loc[(complexity_data_mex_pairs_socioecon['year_pair']==year_pair)&
                                               (complexity_data_mex_pairs_socioecon['Product']==product)]
```

```python
for year_pair in years_pairs:
    
    print(year_pair)

    complexity_data_year=complexity_data_mex_pairs_socioecon.loc[complexity_data_mex_pairs_socioecon['year_pair']==year_pair]
    
    products=sorted(complexity_data_year['Product'].unique())
    
    year_str=year_pair[0:4]
    
    G=nx.read_gexf('../output_data/Graphs_files/'+year_str+".gexf")
    
    Nodes_list=list(G.nodes)
    #Graph attributes
    weights = nx.get_edge_attributes(G,'weight').values()
    # Nodes Metrics
    Nodes_Degree_Centrality_dict = dict(nx.degree_centrality(G))
    Nodes_Eigenvector_Centrality_dict = dict(nx.eigenvector_centrality(G,weight='weight'))
    Nodes_Closeness_Centrality_dict = dict(nx.closeness_centrality(G),distance='distance')
    
    for product in products:
        print(product, end=', ')
        
        complexity_data_mex_pairs_socioecon.loc[(complexity_data_mex_pairs_socioecon['year_pair']==year_pair)&
                                               (complexity_data_mex_pairs_socioecon['Product']==product),
                                                'degree_centrality']=Nodes_Degree_Centrality_dict[str(product)]
        
        complexity_data_mex_pairs_socioecon.loc[(complexity_data_mex_pairs_socioecon['year_pair']==year_pair)&
                                               (complexity_data_mex_pairs_socioecon['Product']==product),
                                                'eigenvecor_centrality']=Nodes_Eigenvector_Centrality_dict[str(product)]
        
        complexity_data_mex_pairs_socioecon.loc[(complexity_data_mex_pairs_socioecon['year_pair']==year_pair)&
                                               (complexity_data_mex_pairs_socioecon['Product']==product),
                                                'closeness_centrality']=Nodes_Closeness_Centrality_dict[str(product)]

        
        
        
complexity_data_mex_pairs_socioecon
```
