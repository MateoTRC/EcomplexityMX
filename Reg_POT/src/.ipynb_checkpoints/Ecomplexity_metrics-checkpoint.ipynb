```python
import pandas as pd
from ecomplexity import ecomplexity
from ecomplexity import proximity
```

```python
# Get dataframes

exports_2004_df = pd.read_excel("../row_data/prox reg_pot 04-19.xlsx", sheet_name='base 2004').set_index('Row Labels').fillna(0)

exports_2009_df = pd.read_excel("../row_data/prox reg_pot 04-19.xlsx", sheet_name='base 2009').set_index('Row Labels').fillna(0)

exports_2014_df = pd.read_excel("../row_data/prox reg_pot 04-19.xlsx", sheet_name='base 2014').set_index('Row Labels').fillna(0)

exports_2019_df = pd.read_excel("../row_data/prox reg_pot 04-19.xlsx", sheet_name='base 2019').set_index('Row Labels').fillna(0)
```

```python
Products_2004 = set(exports_2004_df.columns)
Products_2009 = set(exports_2009_df.columns)
Products_2014 = set(exports_2014_df.columns)
Products_2019 = set(exports_2019_df.columns)

common_products_04_09 = Products_2004 & Products_2009

common_products_04_09 = sorted([x for x in common_products_04_09 if isinstance(x, int)])

common_products_09_14 = Products_2009 & Products_2014

common_products_09_14 = sorted([x for x in common_products_09_14 if isinstance(x, int)])

common_products_14_19 = Products_2014 &  Products_2019

common_products_14_19 = sorted([x for x in common_products_14_19 if isinstance(x, int)])

common_products_04_19 = Products_2004 & Products_2019

common_products_04_19 = sorted([x for x in common_products_04_19 if isinstance(x, int)])

```

```python
locations=list(exports_2004_df.index)[:-1]
locations
```

```python
years=[2004,2009,2014,2019]

year_loc_prod=[]
for year in years:
    if year==2004:
        common_products=Products_2004
        common_products=sorted([x for x in common_products if isinstance(x, int)])
        exports_df=exports_2004_df
    if year==2009:
        common_products=Products_2009
        common_products=sorted([x for x in common_products if isinstance(x, int)])
        exports_df=exports_2009_df
    if year==2014:
        common_products=Products_2014
        common_products=sorted([x for x in common_products if isinstance(x, int)])
        exports_df=exports_2014_df
    if year==2019:
        common_products=Products_2019
        common_products=sorted([x for x in common_products if isinstance(x, int)])
        exports_df=exports_2019_df
        
    for loc in locations:
        
        for prod in common_products:
            
            export=exports_df.loc[loc][prod]
            
            year_loc_prod.append([year,loc,prod,export])
            
data_mex_per_year=pd.DataFrame(year_loc_prod,columns=['year','location_code','hs_product_code','export_value'])

data_mex_per_year
```

```python
data_mex_per_year.to_csv('../output_data/exports_per_year.csv')
```

```python
# Calculate complexity
trade_cols = {'time':'year', 'loc':'location_code', 'prod':'hs_product_code', 'val':'export_value'}
cdata_mex_per_year = ecomplexity(data_mex_per_year, trade_cols)
cdata_mex_per_year
```

```python
cdata_mex_per_year.to_csv('../output_data/complexity_data_mex_per_year.csv')
```

```python
prox_df = proximity(data_mex_per_year, trade_cols)
prox_df
```

```python
prox_df.to_csv('../output_data/proximity_per_year.csv')
```

```python
years=['2004_2009','2009_2014','2014_2019','2004_2019']

year_loc_prod_char=[]
for year in years:
    year1=int(year[0:4])
    year2=int(year[5:])
    
    if year=='2004_2009':
        common_products = common_products_04_09
    if year=='2009_2014':
        common_products = common_products_09_14
    if year=='2014_2019':
        common_products = common_products_14_19
    if year=='2004_2019':
        common_products = common_products_04_19
   
        
    for loc in locations:
        
        for prod in common_products:
            
            eci = cdata_mex_per_year.loc[(cdata_mex_per_year['location_code']==loc)&
                                (cdata_mex_per_year['hs_product_code']==prod)&
                                (cdata_mex_per_year['year']==year1),'eci'].values[0]
            
            pci = cdata_mex_per_year.loc[(cdata_mex_per_year['location_code']==loc)&
                                (cdata_mex_per_year['hs_product_code']==prod)&
                                (cdata_mex_per_year['year']==year1),'pci'].values[0]
            
            density = cdata_mex_per_year.loc[(cdata_mex_per_year['location_code']==loc)&
                                (cdata_mex_per_year['hs_product_code']==prod)&
                                (cdata_mex_per_year['year']==year1),'density'].values[0]
            
            coi = cdata_mex_per_year.loc[(cdata_mex_per_year['location_code']==loc)&
                                (cdata_mex_per_year['hs_product_code']==prod)&
                                (cdata_mex_per_year['year']==year1),'coi'].values[0]
            
            cog = cdata_mex_per_year.loc[(cdata_mex_per_year['location_code']==loc)&
                                (cdata_mex_per_year['hs_product_code']==prod)&
                                (cdata_mex_per_year['year']==year1),'cog'].values[0]
            
            rca1 = cdata_mex_per_year.loc[(cdata_mex_per_year['location_code']==loc)&
                                (cdata_mex_per_year['hs_product_code']==prod)&
                                (cdata_mex_per_year['year']==year1),'rca'].values[0]
            
            rca2 = cdata_mex_per_year.loc[(cdata_mex_per_year['location_code']==loc)&
                                (cdata_mex_per_year['hs_product_code']==prod)&
                                (cdata_mex_per_year['year']==year2),'rca'].values[0]
            
            if rca1<0.5:
                
                if rca2>=1:
                    M='Transition'
                    M_num = 1
                else:
                    M='Underdeveloped'
                    M_num = 0
            else:
                M='NA'
                M_num = 2
                    
            
            year_loc_prod_char.append([year,loc,prod,export,eci,pci,density,coi,cog,rca1,rca2,M,M_num])
            
cdata_mex_pairs=pd.DataFrame(year_loc_prod_char,columns=['year_pair','location_code','Product','export','eci','pci','density','coi','cog','rca1','rca2','M','M_num'])

cdata_mex_pairs
```

```python
cdata_mex_pairs.to_csv('../output_data/complexity_data_mex_pairs.csv')
```

```python
import numpy as np
pob_gdp_region=pd.read_csv(('../../Socioeconomics/pob_gdp_region.csv'))
ied_edos=pd.read_csv(('../../Socioeconomics/ied_edos.csv'))

Region_Entidades_federativas={'Sur':['Campeche', 'Chiapas', 'Guerrero', 'Oaxaca', 'Quintana Roo', 'Tabasco', 'Veracruz', 'YucatÃ¡n'],
'Centro':['Ciudad de MÃ©xico', 'Estado de MÃ©xico', 'Guanajuato', 'Hidalgo', 'Morelos', 'Puebla', 'QuerÃ©taro', 'Tlaxcala'],
'Centro-Norte':['Aguascalientes', 'Baja California Sur', 'Colima', 'Durango', 'Jalisco', 'MichoacÃ¡n', 'Nayarit', 'San Luis PotosÃ­', 'Sinaloa'],
'Norte':['Baja California', 'Chihuahua', 'Coahuila', 'Nuevo LeÃ³n', 'Sonora', 'Tamaulipas', 'Zacatecas']}

reg_year=[]
for reg in list(Region_Entidades_federativas.keys()):
    for year in np.arange(2004,2020):
        reg_year.append([reg,year])
        
ied_region=pd.DataFrame(reg_year,columns=['Region','Year'])

for reg in list(Region_Entidades_federativas.keys()):
    for year in np.arange(2004,2021):
        
        ied=ied_edos.loc[(ied_edos['edo'].isin(Region_Entidades_federativas[reg]))&
                              (ied_edos['anio']==year),'ied'].values.sum()
        
        ied_region.loc[(ied_region['Region']==reg)&
                              (ied_region['Year']==year),'ied']=ied
ied_region
```

```python
cdata_mex_pairs_sc = cdata_mex_pairs.copy()

for year in years:
    for Estado in locations:
        gdp_percap=pob_gdp_region.loc[(pob_gdp_region['Year']==int(year[0:4]))&
                                     (pob_gdp_region['Region']==Estado),'GDPperCAP'].values[0]
        
        ied=ied_region.loc[(ied_region['Year']==int(year[0:4]))&
                            (ied_region['Region']==Estado),'ied'].values[0]
        
        cdata_mex_pairs_sc.loc[(cdata_mex_pairs_sc['location_code']==Estado)&
                                       (cdata_mex_pairs_sc['year_pair']==year),'GPDperCAP']=gdp_percap
        
        cdata_mex_pairs_sc.loc[(cdata_mex_pairs_sc['location_code']==Estado)&
                                       (cdata_mex_pairs_sc['year_pair']==year),'ied']=ied
        
cdata_mex_pairs_sc
```

```python
N=len(locations)
years_pairs=['2004_2009','2009_2014','2014_2019','2004_2019']

for year_pair in years:
    
    if year_pair=='2004_2009':
        common_products = common_products_04_09
    if year_pair=='2009_2014':
        common_products = common_products_09_14
    if year_pair=='2014_2019':
        common_products = common_products_14_19
    if year_pair=='2004_2019':
        common_products = common_products_04_19
        
    for prod in common_products:
        T=len(cdata_mex_pairs_sc.loc[(cdata_mex_pairs_sc['year_pair']==year_pair)&
                          (cdata_mex_pairs_sc['M']=='Transition')&
                          (cdata_mex_pairs_sc['Product']==prod)].values)
        
        M_values=cdata_mex_pairs_sc.loc[(cdata_mex_pairs_sc['year_pair']==year_pair)&
                          (cdata_mex_pairs_sc['Product']==prod),'M'].values
        
        no_trans=[1 if M_val == 'Underdeveloped' else 0 for M_val in M_values]
        
        trans=[1 if M_val == 'Transition' else 0 for M_val in M_values]
        
        dens=cdata_mex_pairs_sc.loc[(cdata_mex_pairs_sc['year_pair']==year_pair)&
                          
                          (cdata_mex_pairs_sc['Product']==prod),'density'].values
        
        H_j=0
        
        if np.dot(dens,no_trans)!=0 and T!=0:
            H_j = (N-T)/T * np.dot(dens,trans)/ np.dot(dens,no_trans)
        else:
            H_j = 0
                
                
        cdata_mex_pairs_sc.loc[(cdata_mex_pairs_sc['year_pair']==year_pair)&
                          (cdata_mex_pairs_sc['Product']==prod),'discovery_factor']=H_j
    
    
cdata_mex_pairs_sc
```
