```python
import pandas as pd # For creating dataframes and calculating correlation matrices.
from pandas.api.types import CategoricalDtype # For correctly setting some dataframe colummns as being of categorical type

from pathlib import Path # For defining the path directory in which the data is stored

import seaborn as sn   # For creating visual representations
import matplotlib.pyplot as plt # of correlation matrices.

import numpy as np #For numerical opertions

import networkx as nx
import matplotlib.cm as cm

```

```python
prox_df=pd.read_csv('../output_data/proximity_per_year.csv')
prox_df['distance']=1-prox_df['proximity']
prox_df
```

```python
cdata_mex_pairs=pd.read_csv('../output_data/complexity_data_mex_pairs.csv')
cdata_mex_pairs
```

```python
Estados=cdata_mex_pairs['location_code'].unique()

years=prox_df['year'].unique()

for year in years:
    
    print(year)

    prox_df_year=prox_df.loc[prox_df['year']==year]
    
    products=sorted(prox_df_year['hs_product_code_1'].unique())
    
    G = nx.Graph()
    
    for prod in products:
        
        print(prod, end=', ')
    
        prox_prod_df=prox_df.loc[prox_df['hs_product_code_1']==prod].loc[prox_df['year']==year].loc[prox_df['proximity']>0]

        prods2=prox_prod_df['hs_product_code_2'].values

        proximity_list=prox_prod_df['proximity'].values

        distance_list=prox_prod_df['distance'].values

        G.add_weighted_edges_from(list((prod, n, w) for n,w in zip(prods2,proximity_list)),weight='proximity')
        G.add_weighted_edges_from(list((prod, n, w) for n,w in zip(prods2,distance_list)),weight='distance')
        
        
    nx.write_gexf(G, '../output_data/Graphs_files/'+str(year)+".gexf")
    print('Finished: ',year)
```
