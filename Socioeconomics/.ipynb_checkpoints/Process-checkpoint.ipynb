```python
import pandas as pd
import numpy as np

# Source:
# Población: https://www.inegi.org.mx/app/tabulados/interactivos/?pxq=Poblacion_Poblacion_01_e60cd8cf-927f-4b94-823e-972457a12d4b
# GDP: https://www.inegi.org.mx/app/tabulados/default.aspx?pr=17&vr=7&in=2&tp=20&wr=1&cno=2
```

```python
poblacion_df=pd.read_excel('Poblacion_01.xlsx')
poblacion_df=poblacion_df.replace('Michoacán de Ocampo','Michoacán')
poblacion_df=poblacion_df.replace('Coahuila de Zaragoza','Coahuila')
poblacion_df=poblacion_df.replace('Veracruz de Ignacio de la Llave','Veracruz')
poblacion_df
```

```python
Estados=['Aguascalientes', 'Baja California', 'Baja California Sur',
       'Campeche', 'Coahuila', 'Colima', 'Chiapas', 'Chihuahua',
       'Ciudad de México', 'Durango', 'Guanajuato', 'Guerrero', 'Hidalgo',
       'Jalisco', 'México', 'Michoacán', 'Morelos', 'Nayarit',
       'Nuevo León', 'Oaxaca', 'Puebla', 'Querétaro', 'Quintana Roo',
       'San Luis Potosí', 'Sinaloa', 'Sonora', 'Tabasco', 'Tamaulipas',
       'Tlaxcala', 'Veracruz', 'Yucatán', 'Zacatecas']

years=[2005,2010,2020]


est_year=[]
for year in years:
    for Estado in Estados:
        pop=poblacion_df.loc[(poblacion_df['Entidad federativa']==Estado)&
                             (poblacion_df['Grupo quinquenal de edad']=='Total'),str(year)].values[0]
        est_year.append([Estado,year,pop])
    
poblacion_clean_df=pd.DataFrame(est_year,columns=['Estado','Year','Population'])

poblacion_clean_df
```

```python
GDP_df=pd.read_csv('PIBE_2.csv')
GDP_df=GDP_df.replace('Michoacán de Ocampo','Michoacán')
GDP_df=GDP_df.replace('Coahuila de Zaragoza','Coahuila')
GDP_df=GDP_df.replace('Veracruz de Ignacio de la Llave','Veracruz')

GDP_df.columns
```

```python
estados_year=[]
for estado in Estados:
    for year in np.arange(2004,2021):
        estados_year.append([estado,year])
        
        

poblacion_interp=pd.DataFrame(estados_year,columns=['Estado','Year'])

poblacion_interp['llave']=poblacion_interp['Estado']+'-'+poblacion_interp['Year'].astype(str)

poblacion_clean_df['llave']=poblacion_clean_df['Estado']+'-'+poblacion_clean_df['Year'].astype(str)


poblacion_interp
```

```python
poblacion_clean_df
```

```python
reune = poblacion_interp.merge(right= poblacion_clean_df[["llave", "Population"]], how = "left", on = "llave")

reune = reune[["Estado", "Year", "Population"]]

reune_interpol = reune.groupby("Estado")["Population"].apply(lambda x : x.interpolate(method='linear', limit_direction='forward', axis=0)).reset_index()


reune_interpol['Estado']=reune['Estado']
reune_interpol['Year']=reune['Year']

reune_interpol

for estado in Estados:
    
    reune_interpol.loc[(reune_interpol['Estado']==estado)&
                      (reune_interpol['Year']==2004),'Population']=reune_interpol.loc[(reune_interpol['Estado']==estado)&
                      (reune_interpol['Year']==2005),'Population'].values[0]
    
poblacion_completo=reune_interpol[['Estado','Year','Population']]

poblacion_completo
```

```python
pob_gdp=poblacion_completo.copy()
pob_gdp['GDP']=0

years_string=[str(y) for y in range(2004,2020)]

for Estado in Estados:
    for year in years_string:
        GDP = GDP_df.loc[GDP_df['Estado']==Estado,year].values[0]
        pob_gdp.loc[(pob_gdp['Estado']==Estado)&
                    (pob_gdp['Year']==int(year)),'GDP']=GDP


pob_gdp = pob_gdp[pob_gdp['GDP'] != 0]

pob_gdp['GDPperCAP']=pob_gdp['GDP']/pob_gdp['Population']

pob_gdp
```

```python
pob_gdp.to_csv('pob_gdp_edos.csv')
```

```python
Region_Entidades_federativas={'Sur':['Campeche', 'Chiapas', 'Guerrero', 'Oaxaca', 'Quintana Roo', 'Tabasco', 'Veracruz', 'Yucatán'],
'Centro':['Ciudad de México', 'Estado de México', 'Guanajuato', 'Hidalgo', 'Morelos', 'Puebla', 'Querétaro', 'Tlaxcala'],
'Centro-Norte':['Aguascalientes', 'Baja California Sur', 'Colima', 'Durango', 'Jalisco', 'Michoacán', 'Nayarit', 'San Luis Potosí', 'Sinaloa'],
'Norte':['Baja California', 'Chihuahua', 'Coahuila', 'Nuevo León', 'Sonora', 'Tamaulipas', 'Zacatecas']}
```

```python
reg_year=[]
for reg in list(Region_Entidades_federativas.keys()):
    for year in np.arange(2004,2020):
        reg_year.append([reg,year])
        
poblacion_por_region=pd.DataFrame(reg_year,columns=['Region','Year'])

for reg in list(Region_Entidades_federativas.keys()):
    for year in np.arange(2004,2021):
        
        pop_reg=pob_gdp.loc[(pob_gdp['Estado'].isin(Region_Entidades_federativas[reg]))&
                              (pob_gdp['Year']==year),'Population'].values.sum()
        
        gdp_reg=pob_gdp.loc[(pob_gdp['Estado'].isin(Region_Entidades_federativas[reg]))&
                              (pob_gdp['Year']==year),'GDP'].values.sum()
        
        poblacion_por_region.loc[(poblacion_por_region['Region']==reg)&
                              (poblacion_por_region['Year']==year),'Population']=pop_reg
        poblacion_por_region.loc[(poblacion_por_region['Region']==reg)&
                              (poblacion_por_region['Year']==year),'GDP']=gdp_reg
        
poblacion_por_region['GDPperCAP']=poblacion_por_region['GDP']/poblacion_por_region['Population']
poblacion_por_region
```

```python
poblacion_por_region.to_csv('pob_gdp_region.csv')
```
