```python
# Import libraries

import pandas as pd
from ecomplexity import ecomplexity
from ecomplexity import proximity
```

```python
# Read input data. 

# Should be disposed as in the excel file prox edo_pot 04-19.xlsx sheet base 2004.



exports_2004_df = pd.read_excel("../row_data/prox edo_pot 04-19.xlsx", sheet_name='base 2004').set_index('Row Labels').fillna(0)

exports_2009_df = pd.read_excel("../row_data/prox edo_pot 04-19.xlsx", sheet_name='base 2009').set_index('Row Labels').fillna(0)

exports_2014_df = pd.read_excel("../row_data/prox edo_pot 04-19.xlsx", sheet_name='base 2014').set_index('Row Labels').fillna(0)

exports_2019_df = pd.read_excel("../row_data/prox edo_pot 04-19.xlsx", sheet_name='base 2019').set_index('Row Labels').fillna(0)
```

```python
# Read products NAICS code

Products_2004 = set(exports_2004_df.columns)
Products_2009 = set(exports_2009_df.columns)
Products_2014 = set(exports_2014_df.columns)
Products_2019 = set(exports_2019_df.columns)

# Get pair of years common products

common_products_04_09 = Products_2004 & Products_2009

common_products_04_09 = sorted([x for x in common_products_04_09 if isinstance(x, int)])

common_products_09_14 = Products_2009 & Products_2014

common_products_09_14 = sorted([x for x in common_products_09_14 if isinstance(x, int)])

common_products_14_19 = Products_2014 &  Products_2019

common_products_14_19 = sorted([x for x in common_products_14_19 if isinstance(x, int)])

common_products_04_19 = Products_2004 & Products_2019

common_products_04_19 = sorted([x for x in common_products_04_19 if isinstance(x, int)])

```

```python
# Get location codes

locations=list(exports_2004_df.index)[:-1]
locations
```

```python
list(sorted(locations)) == locations
```

```python
# Configre data to be processed with ecomplexity library

years=[2004,2009,2014,2019]

year_loc_prod=[]
for year in years:
    if year==2004:
        common_products=Products_2004
        common_products=sorted([x for x in common_products if isinstance(x, int)])
        exports_df=exports_2004_df
    if year==2009:
        common_products=Products_2009
        common_products=sorted([x for x in common_products if isinstance(x, int)])
        exports_df=exports_2009_df
    if year==2014:
        common_products=Products_2014
        common_products=sorted([x for x in common_products if isinstance(x, int)])
        exports_df=exports_2014_df
    if year==2019:
        common_products=Products_2019
        common_products=sorted([x for x in common_products if isinstance(x, int)])
        exports_df=exports_2019_df
        
    for loc in locations:
        
        for prod in common_products:
            
            export=exports_df.loc[loc][prod]
            
            year_loc_prod.append([year,loc,prod,export])
            
data_mex_per_year=pd.DataFrame(year_loc_prod,columns=['year','location_code','hs_product_code','export_value'])

data_mex_per_year
```

```python
# Save ecomplexity configured data

data_mex_per_year.to_csv('../output_data/exports_per_year.csv')
```

```python
# Compute complexity metrix 

trade_cols = {'time':'year', 'loc':'location_code', 'prod':'hs_product_code', 'val':'export_value'}
cdata_mex_per_year = ecomplexity(data_mex_per_year, trade_cols)
cdata_mex_per_year
```

```python
# Save complexity data

cdata_mex_per_year.to_csv('../output_data/complexity_data_mex_per_year.csv')
```

```python
# Compute proximity

prox_df = proximity(data_mex_per_year, trade_cols)
prox_df
```

```python
# Save proximity data

prox_df.to_csv('../output_data/proximity_per_year.csv')
```

```python
# Create dataframe with year pairs data

years=['2004_2009','2009_2014','2014_2019','2004_2019']

years=['2004_2019','2014_2019','2009_2014','2004_2009']


year_loc_prod_char=[]

cdata_mex_pairs=pd.DataFrame(columns=['year_pair','location_code','Product','export','eci','pci','density','coi','cog','rca1','rca2'])

for years_pair in years:
    
    print(years_pair)
    
    cdata_mex_pairs_merge=pd.DataFrame(columns=['year_pair','location_code','Product','export','eci','pci','density','coi','cog','rca1','rca2'])
    
    year1=int(years_pair[0:4])
    year2=int(years_pair[5:])
    
    cdata_mex_year1 = cdata_mex_per_year.loc[cdata_mex_per_year['year'] == year1]
    
    cdata_mex_year2 = cdata_mex_per_year.loc[cdata_mex_per_year['year'] == year2]
    
    common_products = sorted(
    set(cdata_mex_year1['hs_product_code'].values).intersection(set(cdata_mex_year2['hs_product_code'].values)))
        

    cdata_mex_year1 = cdata_mex_year1[cdata_mex_year1['hs_product_code'].isin(common_products)]
    
    cdata_mex_year2 = cdata_mex_year2[cdata_mex_year2['hs_product_code'].isin(common_products)]
    
    prod1 = cdata_mex_year1['hs_product_code'].values
    
    prod2 = cdata_mex_year2['hs_product_code'].values
    
    if not np.unique([prod1 == prod2])[0]:
        print('The products of the first year are not the same as those in the second, or are not in the same order')
        stop
    loc1 = cdata_mex_year1['location_code'].values

    loc2 = cdata_mex_year2['location_code'].values

    if not np.unique([loc1 == loc2])[0]:
            
        print('The products of the first year are not the same as those in the second, or are not in the same order')
        stop
            
    cdata_mex_pairs_merge['Product'] = prod1
    
    cdata_mex_pairs_merge['year_pair'] = years_pair
    
    cdata_mex_pairs_merge['location_code'] = loc1
    
    cdata_mex_pairs_merge['export'] = cdata_mex_year1['export_value'].values
    
    cdata_mex_pairs_merge['eci'] = cdata_mex_year1['eci'].values
    
    cdata_mex_pairs_merge['pci'] = cdata_mex_year1['pci'].values
    
    cdata_mex_pairs_merge['density'] = cdata_mex_year1['density'].values
    
    cdata_mex_pairs_merge['coi'] = cdata_mex_year1['coi'].values
    
    cdata_mex_pairs_merge['cog'] = cdata_mex_year1['cog'].values
    
    cdata_mex_pairs_merge['rca1'] = cdata_mex_year1['rca'].values
    
    cdata_mex_pairs_merge['rca2'] = cdata_mex_year2['rca'].values
    
    cdata_mex_pairs = pd.concat([cdata_mex_pairs_merge,cdata_mex_pairs])
    
cdata_mex_pairs 
```

```python
import numpy as np
pob_gdp_edos=pd.read_csv(('../../Socioeconomics/pob_gdp_edos.csv'))
ied_edos=pd.read_csv(('../../Socioeconomics/ied_edos.csv'))
ied_edos
```

```python
pob_gdp_edos
```

```python
cdata_mex_pairs_sc = cdata_mex_pairs.copy()

for year in years:
    for Estado in locations:
        gdp_percap=pob_gdp_edos.loc[(pob_gdp_edos['Year']==int(year[0:4]))&
                                     (pob_gdp_edos['Estado']==Estado),'GDPperCAP'].values[0]
        
        try:
            ied=ied_edos.loc[(ied_edos['anio']==int(year[0:4]))&
                            (ied_edos['edo']==Estado),'ied'].values[0]
        except:
            ied=0
        
        cdata_mex_pairs_sc.loc[(cdata_mex_pairs_sc['location_code']==Estado)&
                                       (cdata_mex_pairs_sc['year_pair']==year),'GPDperCAP']=gdp_percap
        
        cdata_mex_pairs_sc.loc[(cdata_mex_pairs_sc['location_code']==Estado)&
                                       (cdata_mex_pairs_sc['year_pair']==year),'ied']=ied
        
cdata_mex_pairs_sc
```

```python
N=len(locations)
years_pairs=['2004_2009','2009_2014','2014_2019','2004_2019']

for year_pair in years:
    year1=int(years_pair[0:4])
    year2=int(years_pair[5:])

    cdata_mex_year1 = cdata_mex_per_year.loc[cdata_mex_per_year['year'] == year1]

    cdata_mex_year2 = cdata_mex_per_year.loc[cdata_mex_per_year['year'] == year2]

    common_products = sorted(
    set(cdata_mex_year1['hs_product_code'].values).intersection(set(cdata_mex_year2['hs_product_code'].values)))
        
    for prod in common_products:
        print(prod)
        
        for low_range in low_range_array:
            print(low_range,end=', ')
            T=len(cdata_mex_pairs_sc.loc[(cdata_mex_pairs_sc['year_pair']==year_pair)&
                              (cdata_mex_pairs_sc['M_'+str(low_range).replace('.','_')]=='Transition')&
                              (cdata_mex_pairs_sc['Product']==prod)].values)

            M_values=cdata_mex_pairs_sc.loc[(cdata_mex_pairs_sc['year_pair']==year_pair)&
                              (cdata_mex_pairs_sc['Product']==prod),'M_'+str(low_range).replace('.','_')].values

            no_trans=[1 if M_val == 'Underdeveloped' else 0 for M_val in M_values]

            trans=[1 if M_val == 'Transition' else 0 for M_val in M_values]

            dens=cdata_mex_pairs_sc.loc[(cdata_mex_pairs_sc['year_pair']==year_pair)&

                              (cdata_mex_pairs_sc['Product']==prod),'density'].values
            if np.dot(dens,no_trans)!=0 and T!=0:
                H_j = (N-T)/T * np.dot(dens,trans)/ np.dot(dens,no_trans)
            else:
                H_j = 0
            
            cdata_mex_pairs_sc.loc[(cdata_mex_pairs_sc['year_pair']==year_pair)&
                          (cdata_mex_pairs_sc['Product']==prod),'discovery_factor_'+str(low_range).replace('.','_')]=H_j
    
    
cdata_mex_pairs_sc.to_csv('../output_data/complexity_data_mex_pairs_socioecon.csv')
cdata_mex_pairs_sc
```
