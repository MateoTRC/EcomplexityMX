```python
import pandas as pd
from ecomplexity import ecomplexity
from ecomplexity import proximity
```

```python
# Get dataframes

exports_2004_df = pd.read_excel("../row_data/prox reg_pot 04-19.xlsx", sheet_name='base 2004').set_index('Row Labels').fillna(0)

exports_2009_df = pd.read_excel("../row_data/prox reg_pot 04-19.xlsx", sheet_name='base 2009').set_index('Row Labels').fillna(0)

exports_2014_df = pd.read_excel("../row_data/prox reg_pot 04-19.xlsx", sheet_name='base 2014').set_index('Row Labels').fillna(0)

exports_2019_df = pd.read_excel("../row_data/prox reg_pot 04-19.xlsx", sheet_name='base 2019').set_index('Row Labels').fillna(0)
```

```python
Products_2004 = set(exports_2004_df.columns)
Products_2009 = set(exports_2009_df.columns)
Products_2014 = set(exports_2014_df.columns)
Products_2019 = set(exports_2019_df.columns)

common_products_04_09 = Products_2004 & Products_2009

common_products_04_09 = sorted([x for x in common_products_04_09 if isinstance(x, int)])

common_products_09_14 = Products_2009 & Products_2014

common_products_09_14 = sorted([x for x in common_products_09_14 if isinstance(x, int)])

common_products_14_19 = Products_2014 &  Products_2019

common_products_14_19 = sorted([x for x in common_products_14_19 if isinstance(x, int)])

common_products_04_19 = Products_2004 & Products_2019

common_products_04_19 = sorted([x for x in common_products_04_19 if isinstance(x, int)])

```

```python
locations=list(exports_2004_df.index)[:-1]
locations
```

```python
years=[2004,2009,2014,2019]

year_loc_prod=[]
for year in years:
    if year==2004:
        common_products=Products_2004
        common_products=sorted([x for x in common_products if isinstance(x, int)])
        exports_df=exports_2004_df
    if year==2009:
        common_products=Products_2009
        common_products=sorted([x for x in common_products if isinstance(x, int)])
        exports_df=exports_2009_df
    if year==2014:
        common_products=Products_2014
        common_products=sorted([x for x in common_products if isinstance(x, int)])
        exports_df=exports_2014_df
    if year==2019:
        common_products=Products_2019
        common_products=sorted([x for x in common_products if isinstance(x, int)])
        exports_df=exports_2019_df
        
    for loc in locations:
        
        for prod in common_products:
            
            export=exports_df.loc[loc][prod]
            
            year_loc_prod.append([year,loc,prod,export])
            
data_mex_per_year=pd.DataFrame(year_loc_prod,columns=['year','location_code','hs_product_code','export_value'])

data_mex_per_year
```

```python
data_mex_per_year.to_csv('../output_data/exports_per_year.csv')
```

```python
# Calculate complexity
trade_cols = {'time':'year', 'loc':'location_code', 'prod':'hs_product_code', 'val':'export_value'}
cdata_mex_per_year = ecomplexity(data_mex_per_year, trade_cols)
cdata_mex_per_year
```

```python
cdata_mex_per_year.to_csv('../output_data/complexity_data_mex_per_year.csv')
```

```python
prox_df = proximity(data_mex_per_year, trade_cols)
prox_df
```

```python
prox_df.to_csv('../output_data/proximity_per_year.csv')
```

```python
low_range_array=[0.5,0.55,0.6,0.65,0.7,0.75,0.8,0.85,0.9,0.95,1]

for low_range in low_range_array:

    cdata_mex_pairs['M_num_'+str(low_range).replace('.','_')] = cdata_mex_pairs.apply(lambda row: 1 if row['rca1'] < low_range and row['rca2'] >= 1 else (0 if row['rca1'] < low_range and row['rca2'] < low_range else None), axis=1)

for low_range in low_range_array:

    cdata_mex_pairs['M_'+str(low_range).replace('.','_')] = cdata_mex_pairs.apply(lambda row: 'Transition' if row['M_num_'+str(low_range).replace('.','_')] ==1 else ('Underdeveloped' if row['M_num_'+str(low_range).replace('.','_')] ==0 else 'NA'), axis=1)

cdata_mex_pairs
```

```python
N=len(locations)
years_pairs=['2004_2009','2009_2014','2014_2019','2004_2019']

for year_pair in years:
    if year_pair=='2004_2009':
        common_products = common_products_04_09
    if year_pair=='2009_2014':
        common_products = common_products_09_14
    if year_pair=='2014_2019':
        common_products = common_products_14_19
    if year_pair=='2004_2019':
        common_products = common_products_04_19
    print(year_pair)   
    for prod in common_products:
       
        
        for low_range in low_range_array:
            
            T=len(cdata_mex_pairs_sc.loc[(cdata_mex_pairs_sc['year_pair']==year_pair)&
                              (cdata_mex_pairs_sc['M_'+str(low_range).replace('.','_')]=='Transition')&
                              (cdata_mex_pairs_sc['Product']==prod)].values)

            M_values=cdata_mex_pairs_sc.loc[(cdata_mex_pairs_sc['year_pair']==year_pair)&
                              (cdata_mex_pairs_sc['Product']==prod),'M_'+str(low_range).replace('.','_')].values

            no_trans=[1 if M_val == 'Underdeveloped' else 0 for M_val in M_values]

            trans=[1 if M_val == 'Transition' else 0 for M_val in M_values]

            dens=cdata_mex_pairs_sc.loc[(cdata_mex_pairs_sc['year_pair']==year_pair)&

                              (cdata_mex_pairs_sc['Product']==prod),'density'].values
            if np.dot(dens,no_trans)!=0 and T!=0:
                H_j = (N-T)/T * np.dot(dens,trans)/ np.dot(dens,no_trans)
            else:
                H_j = 0
            
            cdata_mex_pairs_sc.loc[(cdata_mex_pairs_sc['year_pair']==year_pair)&
                          (cdata_mex_pairs_sc['Product']==prod),'discovery_factor_'+str(low_range).replace('.','_')]=H_j
    
    
cdata_mex_pairs_sc.to_csv('../output_data/complexity_data_mex_pairs_socioecon.csv')
cdata_mex_pairs_sc
```
